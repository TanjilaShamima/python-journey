```python
num = input("Enter a number: ")
number = num.split(" ");
number1 = int(number[0]);
number2 = int(number[1]);
print("You entered:", num, "which is of type", type(num))
print("First number:", number1, "which is of type", type(number1))
print("Second number:", number2, "which is of type", type(number2))
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

last_digit_num1 = num1%10
last_digit_num2 = num2%10

if last_digit_num1 == last_digit_num2:
    print("The last digits are the same.")
else:    print("The last digits are different.")

print("Sum of the two numbers:", num1 + num2)
print("Sum of the two last digits:", last_digit_num1 + last_digit_num2)
```
