```python
num = input()
# take num of input and convert it to int
number = input()
number_to_list = number.split()

even = 0
odd = 0
negative = 0
positive = 0
# -5 0 -3 -4 12

for i in range(len(number_to_list)):
    number = number_to_list[i]
    number = int(number)

    if number % 2 == 0:
        even += 1
    else:
        odd += 1

    if number < 0:
        negative += 1
    elif number > 0:
        positive += 1




print("Even: ", even)
print("Odd: ", odd)
print("Positive: ", positive)
print("Negative: ", negative)
```
