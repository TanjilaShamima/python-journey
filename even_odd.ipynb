```python
num = input()
# take num of input and convert it to int
num = int(num)

even = 0
odd = 0
negative = 0
positive = 0

for i in range(num):
    number = input()
    number = int(number)

    if number > 0 and number % 2 == 0:
        even += 1
    else:
        odd += 1

    if number < 0:
        negative += 1
    else:
        positive += 1




print("Even: ", even)
print("Odd: ", odd)
print("Positive: ", positive)
print("Negative: ", negative)
```
